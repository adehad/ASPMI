```python
deep_network_layers = [10,5, 5, 5,1]
epochsA = int(20e1)
epochsB = int(20e3)
epochsC = int(40e3)
epochsD = int(80e3)
learning_rate = 1e-2
noise_powerA = 0.01
noise_powerB = 0.05
noise_powerC = 0.1
noise_powerD = 0.5
```

```python
# consisteny again
random.seed(1)
np.random.seed(1)
torch.manual_seed(1)
```

```python
X = np.load('X.npy')
y = np.load('y.npy') 
yA = y + np.sqrt(noise_powerA)*np.random.randn(y.shape[0])
yB = y + np.sqrt(noise_powerB)*np.random.randn(y.shape[0])
yC = y + np.sqrt(noise_powerC)*np.random.randn(y.shape[0])
yD = y + np.sqrt(noise_powerD)*np.random.randn(y.shape[0])
```

```python
plot_data(X, yA)
plot_data(X, yB)
plot_data(X, yC)
plot_data(X, yD)
```

```python
# What are the models we are teting
models = [LMS(), nonlinear_LMS(), deep_network(deep_network_layers)]
```

```python
#
lossAA = train_models(X, yA, models, epochsA, learning_rate)
lossBA = train_models(X, yB, models, epochsA, learning_rate)
lossCA = train_models(X, yC, models, epochsA, learning_rate)
lossDA = train_models(X, yD, models, epochsA, learning_rate)
```

```python
lossAC = train_models(X, yA, models, epochsC, learning_rate)
lossBC = train_models(X, yB, models, epochsC, learning_rate)
lossCC = train_models(X, yC, models, epochsC, learning_rate)
lossDC = train_models(X, yD, models, epochsC, learning_rate)
```

```python
lossAD = train_models(X, yA, models, epochsD, learning_rate)
lossBD = train_models(X, yB, models, epochsD, learning_rate)
lossCD = train_models(X, yC, models, epochsD, learning_rate)
lossDD = train_models(X, yD, models, epochsD, learning_rate)
```
